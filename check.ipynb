```python
import openai
import os

# Replace with your OpenAI API key
openai.api_key = os.getenv("OPENAI_API_KEY")  # or directly input key

def test_openai_key():
    try:
        # Test if OpenAI API key works
        response = openai.Completion.create(
            model="gpt-4",  # Use the model you have access to
            prompt="Test API key",
            max_tokens=5
        )
        print("OpenAI API key works:", response)
        return True
    except Exception as e:
        print("Error with OpenAI API key:", e)
        return False
```

```python
import httpx
import os

tavily_key = os.getenv("TAVILY_API_KEY")  # or directly input key

def test_tavily_key():
    try:
        response = httpx.post(
            "https://api.tavily.com/search",
            json={
                "api_key": tavily_key,
                "query": "test",
                "max_results": 3,
                "search_depth": "basic",
                "include_answer": True,
            },
            timeout=8,
        )
        data = response.json()
        print("Tavily API key works:", data)
        return True
    except Exception as e:
        print("Error with Tavily API key:", e)
        return False
```

```python
from arize.otel import register
import os

def test_arize_key():
    try:
        tracer_provider = register(
            space_id=os.getenv("ARIZE_SPACE_ID"),
            api_key=os.getenv("ARIZE_API_KEY"),
            project_name="test-project",
            set_global_tracer_provider=True
        )
        print("Arize API key works")
        return True
    except Exception as e:
        print("Error with Arize API key:", e)
        return False
```

```python
def check_api_keys():
    if test_openai_key():
        print("OpenAI API key is valid!")
    else:
        print("OpenAI API key is invalid!")

    if test_tavily_key():
        print("Tavily API key is valid!")
    else:
        print("Tavily API key is invalid!")

    if test_arize_key():
        print("Arize API key is valid!")
    else:
        print("Arize API key is invalid!")

check_api_keys()
```
